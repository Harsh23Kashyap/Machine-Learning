# **Air Quality Prediction Using In built Library**

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
model= LinearRegression()
```

```python
Train=pd.read_csv("Train.csv")
Test_X=pd.read_csv("Test.csv")
```

```python
Train
```

```python
Train_X=Train.iloc[:,0:-1]
Train_Y=Train.iloc[:,-1]
```

```python
Train_X=np.array(Train_X)
Train_Y=np.array(Train_Y)
Test_X=np.array(Test_X)
Test_Y=np.array(Test_Y)
```

```python
def Normaliser(X):
  m,n=X.shape
  for i in range(n):
    mean=np.mean(X[:,i])
    std=np.std(X[:,i])
    for j in range(m):
      X[j,i]=(X[j,i]-mean)/std
  return X
```

```python
#Normalise datset
Train_X=Normaliser(Train_X)
```

```python
Train_X[:,0].shape, Train_Y.shape
```

```python
plt.style.use('seaborn')
plt.subplot(1,2,1)
plt.scatter(Train_X[:,0],Train_Y,color='red')
plt.title("Feature 1 vs Result")
plt.xlabel("Feature 1")
plt.ylabel("Result")
plt.show()

plt.subplot(1,2,2)
plt.scatter(Train_X[:,1],Train_Y,color='green')
plt.title("Feature 2 vs Result")
plt.xlabel("Feature 2")
plt.ylabel("Result")
plt.show()

plt.subplot(1,2,2)
plt.scatter(Train_X[:,2],Train_Y,color='blue')
plt.title("Feature 3 vs Result")
plt.xlabel("Feature 3")
plt.ylabel("Result")
plt.show()

plt.subplot(1,3,2)
plt.scatter(Train_X[:,3],Train_Y,color='brown')
plt.title("Feature 4 vs Result")
plt.xlabel("Feature 4")
plt.ylabel("Result")
plt.show()

plt.subplot(2,3,2)
plt.scatter(Train_X[:,4],Train_Y,color='yellow')
plt.title("Feature 5 vs Result")
plt.xlabel("Feature 5")
plt.ylabel("Result")
plt.show()
```

```python
model.fit(Train_X,Train_Y)
```

```python
print("Score of the Model is "+str(np.round((model.score(Train_X,Train_Y)),2)*100)+" %")
```
