```python
class Prop:
    def __init__(self):
        self.sq=int(input("enter square footage"))
        self.room=int(input("enter bedrooms"))
        self.bath=int(input("enter bathroom"))
    
class House(Prop):
    def __init__(self):
        super().__init__()
        self.stories=int(input("Enter no of stories"))
        self.garage=input("is garage attached")
        self.yard=input("is yard fenced")
    
class Apar(Prop):
    def __init__(self):
        super().__init__()
        self.balcony=input("is balcony available")
        self.laundry=input("laundry status")

class Rent:
    def __int__(self):
        super().__init__()
        self.price=int(input("enter rent"))
        self.furnished=input("is furnished")
        self.util=input("is utility included")
    

class Purch:
    def __init__(self):
        super().__init__()
        self.price=int(input("enter price"))
        self.tax=int(input("enter app. tax"))
    
class A1(Apar,Purch):
    types="Apartment"
    val="Purchase"
    status="Available"
    def display(self):
        print("type ",self.types," Listed as",self.val," current status ",self.status,"\n")
        print(" square footage",self.sq," no. of bed rooms",self.room," no. of bathroom", self.bath,"\n")
        print("balcony avail.",self.balcony," laundry status ",self.laundry,"\n")
        print("Price ",self.price,"Property tax appx. ",self.tax,"\n\n")
    
class A2(Apar,Rent):
    types="Apartment"
    val="Rent"
    status="Available"
    def display(self):
        print("type ",self.types," Listed as",self.val," current status ",self.status,"\n")
        print(" square footage",self.sq," no. of bed rooms",self.room," no. of bathroom", self.bath,"\n")
        print("balcony avail.",self.balcony," laundry status ",self.laundry,"\n")
        print("Rent amount",self.price," Furnished status ",self.furnished," utility status",self.util,"\n\n")
    
class H1(House,Purch):
    types="House"
    val="Purchase"
    status="Available"
    def display(self):
        print("type ",self.types," Listed as",self.val," current status ",self.status,"\n")
        print(" square footage",self.sq," no. of bed rooms",self.room," no. of bathroom", self.bath,"\n")
        print("No of stories ", self.stories," Garage status ",self.garage," Is yard fenced ",self.yard,"\n")
        print("Price ",self.price,"Property tax appx. ",self.tax,"\n\n")
        
        
    
class H2(House,Rent):
    super(House,self).__init__()
    super(Rent,self).__init__()
    types="House"
    val="Rent"
    status="Available"
    def display(self):
        print("type ",self.types," Listed as",self.val," current status ",self.status,"\n")
        print(" square footage",self.sq," no. of bed rooms",self.room," no. of bathroom", self.bath,"\n")
        print("No of stories ", self.stories," Garage status ",self.garage," Is yard fenced ",self.yard,"\n")
        print("Rent amount",self.price," Furnished status ",self.furnished," utility status",self.util,"\n\n")
          
        
class Agent:
    instances=[]
    def create(self):
        a=input("Enter property type")
        b=input("Enter Rental or Purchase")
        if a=="House" and b=="Rental":
            ob = H2() 
            self.instances.append(ob)
        elif a=="House" and b=="Purchase":
            ob = H1() 
            self.instances.append(ob)
        elif a=="Apartment" and b=="Rental":
            ob = A2() 
            self.instances.append(ob)
        else:
            ob = A1() 
            self.instances.append(ob)
            
    def disp(self):
        for i in self.instances:
            i.display()
        
A=Agent()
A.create()
A.create()
A.disp()

    
```
