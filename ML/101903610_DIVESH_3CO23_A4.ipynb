```python
# Create a CSV file to store the roll number and marks of five subjects of the students. 
# Find the mean of the marks of each student as well as total mean of the data stored. 
# Select the best case, average case and worst case using the means calculated. 
# Display the roll number of these three types of students.
```

```python
import os
os.getcwd()
os.remove("student_marks.csv")
```

```python
marks_range = range(1,100)
subjects = 5
num_students = 30 
```

```python

import csv
import random

with open('student_marks.csv', 'x') as f: 
    writer = csv.writer(f)
    
    for student in range(num_students):
        student_marksheet = random.sample(marks_range, subjects) 
        writer.writerow(student_marksheet)
```

```python
markshit = np.genfromtxt('student_marks.csv', delimiter=',', dtype = 'int')
```

```python
markshit 
```

```python
mean_each_student = list(map(lambda student_marks: sum(student_marks)/subjects, markshit)) 
```

```python
np.mean(mean_each_student)  
```

```python
print('The mean marks are :' ,np.mean(markshit))
```

```python
worst_case, avg_case, best_case = min(mean_each_student), np.mean(mean_each_student), max(mean_each_student)
```

```python
np.std(mean_each_student) 
```

```python
neighbs = abs(np.array(mean_each_student) - avg_case)
nearest_avg_case = np.argmin(neighbs)
```

```python
print('The worst student award goes to : ', np.where(mean_each_student == worst_case)[0][0])
print('The most average student is : ', nearest_avg_case)
print('The best student award goes to : ', np.where(mean_each_student == best_case)[0][0])
```

```python
# Generate a random sequence and 
# then pick up the samples from the CSV file stored in iris dataset. 
# Then find the mean and standard deviation of these randomly selected samples.
```

```python
rows_iris = len(iris) 
```

```python
random_rows = random.sample(range(0, rows_iris), 10) 
```

```python
random_rows
```

```python
random_iris = iris[random_rows]
```

```python
features = ['Sepal_length', 'Sepal_width', 'Petal_length', 'Petal_width']
for column in range(iris.shape[1]):
    mean = np.mean(random_iris[:, column])
    std = np.std(random_iris[:, column])
    print( str(features[column]) )  
    print('Mean : ', mean)
    print('Standard Deviation : ', std)
```
