```python
#Importing Libraries and Full Body Detection Files
import cv2
detector=cv2.CascadeClassifier('haarcascade_fullbody.xml')
```

```python
#Read the file 
img = cv2.imread("photo.jpeg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
Id="detectedSample"
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
while(True):
    body = detector.detectMultiScale(img, 1.1, 4)
    for (x,y,w,h) in body:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,130,20),2)
        #saving the captured body into directory
        filename=r"Body"+Id +".jpeg"
        cv2.imwrite(filename, img[y:y+h,x:x+w])
        #showing the frame with a rectangle surrounding the body
        cv2.imshow('frame',img)

    #press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
img = cv2.imread(filename)
plt.imshow(img)
```
